```python
#Find all records where the weather was exactly clear
clear_weather = df[df["Weather"]=="Clear"]
print(clear_weather)
```

```python
#Filter records where the wind speed was exactly 4 km/hr
wind_speed_4_kmh_records = df[df["Wind Speed_km/h"]==4]

#Count the number of the records
count_wind_speed_4_kmh = len(wind_speed_4_kmh_records)

#Display the count
print(wind_speed_4_kmh_records)
print(f"The number of times the wind speed was exactly 4 km/h is: {count_wind_speed_4_kmh}")
```

```python
#Check if there are any NULL values present in the dataset.
df.isnull().any()
```

```python
#What is the mean visibility of the dataset?
df["Visibility_km"].mean()
```

```python
#Rename the column "Weather" to "Weather_Condition"
# Rename the column
df.rename(columns={"Weather": "Weather_Condition"}, inplace=True)

# Print the DataFrame to verify the change
print("\nDataFrame after renaming column:")
print(df)
```

```python
#Find the number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km.
wind_speed_visibility = df[(df["Wind Speed_km/h"] > 24) & (df["Visibility_km"]==25)]

#Count the number of the records
count_wind_speed_visibility = len(wind_speed_visibility)

#Display the count
print(wind_speed_visibility)
print(f"The number of records the wind speed is greater than 24 km/hr and visibility is equal to 25 km is: {count_wind_speed_visibility}")
```

```python
#The mean value of each column for each weather condition
df["Date/Time"] = pd.to_datetime(df["Date/Time"])
mean_values = df.groupby("Weather_Condition").mean()
print(mean_values)
```

```python
#Find all instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40.
filtered_df = df[(df["Weather_Condition"] == "Clear") & ((df["Rel Hum_%"] > 50) | (df["Visibility_km"] > 40))]
print(filtered_df)
```

```python
#Find the number of weather conditions that include snow.
#Filter for weather conditions including "Snow".
snow_conditions = df[df["Weather_Condition"].str.contains("Snow")]

#Find unique weather conditions that include "Snow"
unique_snow_conditions = snow_conditions["Weather_Condition"].unique()

#Count the number of unique weather conditions
num_unique_snow_conditions = len(unique_snow_conditions)

print(num_unique_snow_conditions)
```
